```python
from bs4 import BeautifulSoup as Soup
from urllib.request import urlopen as uReq
import requests
```

```python
url = "https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_0ff7ac30-1ae3-487e-a54a-b8aadad39c82_2_372UD5BXDFYS_MC.34WHNYFH5V2Y&otracker=hp_rich_navigation_8_2.navigationCard.RICH_NAVIGATION_Electronics~Laptop%2Band%2BDesktop_34WHNYFH5V2Y&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_8_L1_view-all&cid=34WHNYFH5V2Y"
resp = requests.get(url)
print(resp.text)
```

```python
s = Soup(resp.content,features='lxml')
s.select('._4rR01T')[0].get_text() ##Product Name
```

```python
s.select('._1_WHN1')[0].get_text() ##Product Price
```

```python
s.select('._3LWZlK')[0].get_text() ##Product Rating
```

```python
fname = "Flipkart Laptop Data.csv"
f = open(fname,'w')
```

```python
headers = "Product_Name, Price, Rating \n"
f.write(headers)
```

```python
for i in range(0,20):
    name = s.select('._4rR01T')[i].get_text()
    price = s.select('._1_WHN1')[i].get_text()
    rating = s.select('._3LWZlK')[i].get_text()
    print("Product Name:" + s.select('._4rR01T')[0].get_text() )
    print("Price:" + s.select('._1_WHN1')[0].get_text())
    print("Rating:" + s.select('._3LWZlK')[0].get_text())
    print(name.replace(","," |") + "," + price.replace("₹","Rs.").replace(",","") + "," +   rating + "\n")
    f.write(name.replace(","," |") + "," + price.replace("₹","Rs.").replace(",","") + "," +   rating + "\n")
f.close()    
```
